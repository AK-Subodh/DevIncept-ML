### Visualization

Distribution

```python
ird.hist(bins=50,figsize=(12,9))
plt.show()
```

Plotting two attributes in combinations

```python
fig,axes=plt.subplots(2,3,figsize=(15,10))
fig.suptitle("Variety")
sns.set_style("whitegrid")
sns.scatterplot(data=ird,ax=axes[0,0],hue ="species",x='sepal_length',y='petal_length')
sns.scatterplot(data=ird,ax=axes[0,1],hue ="species",x='sepal_length',y='petal_width')
sns.scatterplot(data=ird,ax=axes[0,2],hue ="species",x='sepal_length',y='sepal_width')
sns.scatterplot(data=ird,ax=axes[1,0],hue ="species",x='sepal_width',y='petal_length')
sns.scatterplot(data=ird,ax=axes[1,1],hue ="species",x='sepal_width',y='petal_width')
sns.scatterplot(data=ird,ax=axes[1,2],hue ="species",x='sepal_length',y='petal_width')
```

Checkinf correlation between attributes using heatmap

```python
#checking correlation between variables
fig = plt.figure(figsize = (12,10))
sns.heatmap(ird.corr(), cmap='Blues', annot = True) 
```
